```python
import os
import pandas as pd
import pymysql
```

```python
# 데이터 프레임 작게 보기
pd.set_option('display.max_rows', 10) # 최대 행 개수
pd.set_option('display.width', 1000) # 화면 너비
```

```python
# 재료1과2.csv 파일을 읽어온다
df = pd.read_csv('재료1과2.csv', encoding='cp949', index_col=0)
df
```

```python
# 재료3.csv 파일을 읽어온다
df2 = pd.read_csv('재료3.csv', encoding='cp949', index_col=0)
df2
```

```python
# df와 df2를 합친다
df3 = pd.concat([df, df2], axis=1)
df3
```

```python
# df3 데이터 프레임 전체 행을 
df3.head(44)
```

```python
# df3 ingredient3_2 컬럼의 nan 값을 ''로 바꾼다
df3['ingredient3_2'] = df3['ingredient3_2'].fillna('-1')
df3
```

```python
# df3 ingredient3_2 컬럼을 인트로 만든다
df3['ingredient3_2'] = df3['ingredient3_2'].astype(int)
df3
```

```python
df3_index = df3.index
```

```python
# df3_index를 리스트로 만든다
df3_index = df3_index.tolist()
df3_index
```

```python
# 첫번째 행의 인덱스를 가져온다
df3_index[0]
# 첫번째 행의 ingredient1_2	ingredient2_2	ingredient3_2를 가져온다
df3.loc[df3_index[0], ['ingredient1_2', 'ingredient2_2', 'ingredient3_2']]
# 첫번째 행의 ingredient1_2	ingredient2_2	ingredient3_2를 리스트로 만든다
df3_0_loc_0 = df3.loc[df3_index[0], ['ingredient1_2', 'ingredient2_2', 'ingredient3_2']].tolist()
```

```python
# 부재료 - 시트1.csv 파일을 판다스로 읽어옵니다
df5 = pd.read_csv('부재료 - 시트1.csv', encoding='utf-8')


# df2 index를 1부터 시작하게 바꿔줍니다
df5.index = range(1, len(df5) + 1)
df5
```

```python
df5 = df5["재료"]
# df2를 데이터프레임으로 바꿔줍니다
df5 = pd.DataFrame(df5)
df5

# df2를 리스트로 바꿔줍니다
df5_list = df5.values.tolist()
```

```python
conn = pymysql.connect(host=Config.HOST, user=Config.DB_USER, password=Config.DB_PASSWORD, db=Config.DATABASE, charset='utf8')
curs = conn.cursor()

for i in range(len(df3_index)):
    print(str(i) +" 번째 행입니다.")
    for j in range(len(df5_list)):
        # print(df5_list[j] + " 이 재료가 있나요?")
        if j in df3.loc[df3_index[i], ['ingredient1_2', 'ingredient2_2', 'ingredient3_2']].tolist():
            print("레시피 " + str(i) + " 번째 에는 " + df5_list[j] +" 가 있습니다")
            sql = "insert into recipeIngredient (recipeId , ingredientId) values (%s, %s)"
            curs.execute(sql, (i, j))
            print("저장되었습니다")
            
conn.commit()
conn.close()
curs.close()
        
            
```
