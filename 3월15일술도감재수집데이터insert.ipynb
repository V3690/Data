```python
# 언네임드 컬럼 삭제
df.drop('Unnamed: 0', axis=1, inplace=True)
df.head()
```

```python
# percent 컬럼에서 % 제거
df['percent'] = df['percent'].str.replace('%', '')
df.head()
```

```python
# percent를 float형으로 변환
df['percent'] = df['percent'].astype(float)
df.info()
```

```python
# 데이터베이스 연결
conn = pymysql.connect(host=Config.HOST, user=Config.DB_USER, password=Config.DB_PASSWORD, db=Config.DATABASE, charset='utf8')
curs = conn.cursor()
try:
    for i in range(len(df)):
        sql = "INSERT INTO alcohol (name, percent, alcoholType,category,produce,supply , imgUrl) VALUES (%s, %s, %s,%s,%s,%s, %s)"
        curs.execute(sql, (df['name'][i], df['percent'][i], df['class'][i], df['category'][i], df['produce'][i], df['supply'][i], df['imgUrl'][i]))

except:
        conn.rollback()
        print("예기치 않은 오류:", sys.exc_info()[0])  
        raise 

finally:
    conn.commit()
    conn.close()
    curs.close()
```

```python
import requests
from io import BytesIO
from PIL import Image
import os
import sys
import shutil
import urllib.request
import urllib.parse
import urllib.error
import ssl
import re
import time
import random
import json
import pandas as pd
import numpy as np
# 이미지 다운로드
def download_image(url, file_name):
    try:
        # urlretrieve() 함수로 이미지 다운로드
        urllib.request.urlretrieve(url, file_name)  
        print('다운로드 완료')
    except Exception as e:
        print(e)
        print('다운로드 실패')
```
