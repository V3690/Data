```python
# 칵테일목록.csv파일을 판다스로 읽어옵니다

import pandas as pd
```

```python
df = pd.read_csv('칵테일목록.csv', encoding='utf-8')
# nan값을 공백으로 바꿔줍니다
df = df.fillna('')
df
# -1 ~ -3번째 열의 데이터를 float에서 int로 변환합니다

# 만약 빈값이 있다면 해당 값은 0으로 바꿔줍니다
df = df.replace('', 0)
df
df.iloc[:, -1] = df.iloc[:, -1].astype(int)
df.iloc[:, -2] = df.iloc[:, -2].astype(int)
# -1 ~ -3번째 열의 데이터를 str로 변환하고 : 을 추가해서 새로운 열을 만듭니다
# df['재료'] = df.iloc[:, -1] + ':' + df.iloc[:, -2] + ':' + df.iloc[:, -3]
# df['재료'] = df.iloc[:, -3].astype(str) + ':' + df.iloc[:, -2].astype(str) + ':' + df.iloc[:, -1].astype(str)

```

```python
# df에 0이 있다면 공백으로 바꿔줍니다
df = df.replace(0, '')
df
df['비율'] = "비율 " + df.iloc[:, -3].astype(str) + ':' + df.iloc[:, -2].astype(str) + ':' + df.iloc[:, -1].astype(str)
```

```python
# df["비율"]의 값 맨뒤에 :이 있으면 공백으로 바꿔줍니다
df['비율'] = df['비율'].str.rstrip(':')
df.head()
```

```python
df['레시피임시'] = df['ingredient1'] + ', ' + df['ingredient2'] + ', ' + df['ingredient3']
df['레시피임시'] = df["레시피임시"].str.rstrip(', ')
df
```

```python
df["비율"] = df["비율"].str.replace('비율 ', '')
df
```

```python
# df["ingredient1"] df["ingredient2"] df["ingredient3"] df["비율"]을 하나로 합친다
df['레시피'] = df['레시피임시'] + '를 ' + df['비율'] + " 비율로 섞어서 만들어요"
df
```

```python
# df 가공이 끝났으니 csv파일로 저장합니다
df.to_csv('칵테일목록가공완료.csv', encoding='utf-8-sig', index=False)
```
